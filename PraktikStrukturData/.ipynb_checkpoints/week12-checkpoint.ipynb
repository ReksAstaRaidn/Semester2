# Membuat Dictionary

```python
# 1. membuat dictionary 
my_dict = {}
print(f"Dictionary kosong: {my_dict}")

# dictionary dengan kata awal 
student_info = {
    "nama": "Budi",
    "umur": "20",
    "jurusan": "Informatika",
    "nilai_rata": 3.85
}
print(f"Dictionary student_info: {student_info}")
```

# Mengakses Nilai

```python
# akses nilai menggunakan kunci didalam tanda kurung siku []

print(f"Nama mahasiswa: {student_info['nama']}")
print(f"Jurusan mahasiswa: {student_info['jurusan']}")

#Menggunakan metode .get() untuk menghindari KeyError jika kunci tidak ada 
print(f"Nilai rata-rata: {student_info.get('nilai_rata', 'Tidak ada data')}")
print(f"Tahun masuk (menggunakan .get()): {student_info.get('tahun_masuk', 'Tidak ada data')}")
```

# Menambahkan atau Memperbarui nilai

```python
# dapat menambahkan pasangan kunci-nilai baru atau memeperbarui yang ada dengan sintaks

# menambahkan kunci baru 
student_info['universitas'] = 'Universitas X'
print(f"Setelah menambahkan universitas: {student_info}")

# Memperbarui nilai yang sudah ada 
student_info['umur'] = 21 
print(f"Setelah memperbarui umur: {student_info}")
```

# Menghapus Pasangan kunci-nilai

```python
# hapus pasangan kunci-nilai dengan 'del' atau methode 'pop()'

# menghapus del
del student_info['nilai_rata']
print(f"Setelah menghapus nilai_rata dengan del: {student_info}")

# menghapus dengan .pop() (mengembalikan nilai yang dihapus)
jurusan_dihapus = student_info.pop('jurusan')
print(f"Jurusan yang dihapus: {jurusan_dihapus}")
print(f"Setelah menghapus jurusan dengan pop: {student_info}")

# .pop() juga bisa menerima nilai default jika kunci tidak ditemukan 
no_such_key = student_info.pop('tidak_ada', 'Kunci tidak ditemukan')
print(f"Mencoba menghapus kunci yang tidak ada: {no_such_key}")
```

# Iterasi (Mengulang) melalui dictionary 

```python
# bisa mengulang melalui kunci, nilai, atau keduanya

print("\nIterasi kunci:")
for key in student_info:
    print(f"Kunci: {key}")

print("\nIterasi nilai:")
for value in student_info.values():
    print(f"Nilai: {value}")

print("\nIterasi kunci dan nilai:")
for key, value in student_info.items():
    print(f"{key}: {value}")    
```

# Permasalahan: Menghitung Frekuensi Kata

```python
# Kita memiliki daftar kata dan ingin mengetahui berapa kali setiap kata muncul.

# Contoh data
kalimat = "ini adalah sebuah contoh kalimat ini adalah kalimat yang berulang"
kata_kata = kalimat.split()

print(f"Daftar kata: {kata_kata}")

# Menggunakan Dictionary (Hash Table) untuk menyimpan frekuensi
frekuensi_kata = {}

for kata in kata_kata:
    # Jika kata sudah ada di dictionary, tambahkan 1 ke nilainya
    # Jika belum ada, inisialisasi nilainya dengan 1
    frekuensi_kata[kata] = frekuensi_kata.get(kata, 0) + 1

print("\nFrekuensi kata:")
for kata, jumlah in frekuensi_kata.items():
    print(f"Kata '{kata}': {jumlah} kali")

# Mengapa Hash Function (Dictionary) Efisien di Sini?
# 1. Pencarian Cepat: Untuk setiap kata, dictionary dapat dengan cepat memeriksa
# apakah kata tersebut sudah ada sebagai kunci dan mengambil nilainya (O(1) rata-rata).
# 2. Penambahan/Pembaruan Cepat: Menambahkan kata baru atau memperbarui jumlah kata
# yang sudah ada juga sangat cepat (O(1) rata-rata).
# Jika kita menggunakan daftar biasa dan mencari setiap kata secara sekuensial,
# kompleksitasnya akan jauh lebih tinggi (O(N) untuk setiap pencarian, total O(N*M)).
```

# Implementasi Sederhana Hash Table

```python
class HashTable:
    def __init__ (self, size):
        self.size = size
        self.table = [[] for _ in range(self.size)] # Inisialisasi bucket dengan list kosong
    
    def _hash(self, key):
        # Fungsi hash sederhana: jumlahkan nilai ASCII karakter dan modulus dengan ukuran table
        return sum(ord(char) for char in str(key)) % self.size
    
    def insert(self, key, value):
        index = self._hash(key)
        # Cek apakah kunci sudah ada di bucket tersebut (untuk update atau menghindari duplikat)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value) # Perbarui nilai jika kunci sudah ada
                return
        self.table[index].append((key, value)) # Tambahkan pasangan kunci-nilai baru
        print(f"Menambahkan '{key}': '{value}' di indeks {index}")

    def get(self, key):
        index = self._hash(key)
        for k, v in self.table[index]:
            if k == key:
                return v
            return None # Mengembalikan None jika kunci tidak ditemukan

    def delete(self, key):
        index = self._hash(key)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                del self.table[index][i]
                print(f"Menghapus '{key}' dari indeks {index}")
                return True
        print(f"Kunci '{key}' tidak ditemukan untuk dihapus.")
        return False
        
    def display(self):
        print("\n--- Isi Hash Table ---")
        for i, bucket in enumerate(self.table):
            print(f"Bucket {i}: {bucket}")
        print(" ------------------------------------ ")

# --- Demo Penggunaan Hash Table Kustom ---

# Membuat hash table dengan ukuran 10 bucket
my_custom_hash_table = HashTable(size=10)

# Menambahkan beberapa pasangan kunci-nilai
my_custom_hash_table.insert("apple", 10)
my_custom_hash_table.insert("banana", 20)
my_custom_hash_table.insert("cherry", 30)
my_custom_hash_table.insert("date", 40)
my_custom_hash_table.insert("apple", 15) # Mengupdate nilai 'apple'
my_custom_hash_table.insert("elderberry", 50) # Mungkin akan terjadi kolisi dengan kunci lain

my_custom_hash_table.display()

# Mengambil nilai
print(f"Nilai 'banana': {my_custom_hash_table.get('banana')}")
print(f"Nilai 'apple': {my_custom_hash_table.get('apple')}") # Mengambil nilai yang sudah diupdat
print(f"Nilai 'grape': {my_custom_hash_table.get('grape')}") # Kunci tidak ada

# Menghapus nilai
my_custom_hash_table.delete("cherry")
my_custom_hash_table.delete("grape") # Mencoba menghapus kunci yang tidak ada

my_custom_hash_table.display()

# Contoh kolisi (jika terjadi dengan fungsi hash sederhana ini)
# Misal 'elderberry' dan 'date' mungkin memiliki hash yang sama
print("\n--- Menguji Kolisi ")
my_custom_hash_table.insert("lemon", 60)
my_custom_hash_table.insert("melon", 70)
my_custom_hash_table.display()
```
