# Stack: Convert Integer to Binary #

```python
class Stack():
    def __init__ (self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def is_empty(self):
        return self.items == []

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
    
    def get_stack(self):
        return self.items

s = Stack()

s.push("A")
s.push("B")
print(s.get_stack())
s.push("C")
print(s.get_stack())
s.pop()
print(s.get_stack())
```

```python
s = Stack()
print(s.is_empty())
s.push("A")
print(s.is_empty())
```

```python
s = Stack()

s.push("A")
s.push("B")
s.push("C")
s.push("D")

print(s.peek())
```

# Stack: Balanced Parens

```python
# (), ()(), (({{[]}})) = Balanced 
# ((), {{{)}], [][]]] = Not Balanced 

def is_match(p1, p2):
    if p1 == "(" and p2 == ")":
        return True 
    elif p1 == "{" and p2 == "}":
        return True 
    elif p1 == "[" and p2 == "]":
        return True 
    else: 
        return False 
def is_paren_balanced(paren_string):
    s = Stack()
    is_balanced = True 
    index = 0

    while index < len(paren_string) and is_balanced:
        paren = paren_string[index] 
        if paren in "({[":
            s.push(paren)
        else:
            if s.is_empty():
                is_balanced = False 
            else:
                top = s.pop()
                if not is_match(top, paren):
                    is_balanced = False 
            
        index += 1

    if s.is_empty() and is_balanced:
        return True 
    else:
        return False 
    
print(is_paren_balanced("()")) 
print(is_paren_balanced("(((({}))))")) 

print(is_paren_balanced("[][]]]"))

    
```

# Stack: Divide by Two

```python
# contoh 242 

# convert int to binary

# 242 / 2 = 121 -> 0
# 121 / 2 = 60,5
# 121 / 2 = 60 -> 1
# 60 / 2 = 30 -> 0
# 30 / 2 = 15 -> 0
# 15 / 2 = 7 -> 1
# 7 / 2 = 3 -> 1
# 3 / 2 = 1 -> 1
# 1 / 2 = 0 -> 1

def div_by_2(dec_num):
    s = Stack()

    while dec_num > 0:
        remainder = dec_num % 2 
        s.push(remainder)
        dec_num = dec_num // 2 

    bin_num = ""
    while not s.is_empty():
        bin_num += str(s.pop())
    
    return bin_num

print(div_by_2(242))
    
```
