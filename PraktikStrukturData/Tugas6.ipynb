```python
class HashTable:
    def __init__ (self, size):
        self.size = size
        self.table = [[] for _ in range(self.size)] # Inisialisasi bucket dengan list kosong
    
    def _hash(self, key):
        # Fungsi hash sederhana: jumlahkan nilai ASCII karakter dan modulus dengan ukuran table
        return sum(ord(char) for char in str(key)) % self.size
    
    def insert(self, key, value):
        index = self._hash(key)
        # Cek apakah kunci sudah ada di bucket tersebut (untuk update atau menghindari duplikat)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value) # Perbarui nilai jika kunci sudah ada
                return
        self.table[index].append((key, value)) # Tambahkan pasangan kunci-nilai baru
        print(f"Menambahkan '{key}': '{value}' di indeks {index}")

    def get(self, key):
        index = self._hash(key)
        for k, v in self.table[index]:
            if k == key:
                return v
            return None # Mengembalikan None jika kunci tidak ditemukan

    def delete(self, key):
        index = self._hash(key)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                del self.table[index][i]
                print(f"Menghapus '{key}' dari indeks {index}")
                return True
        print(f"Kunci '{key}' tidak ditemukan untuk dihapus.")
        return False
        
    def display(self):
        print("\n--- Isi Hash Table ---")
        for i, bucket in enumerate(self.table):
            print(f"Bucket {i}: {bucket}")
        print(" ------------------------------------ ")
        
def contains_duplicate(nums):
    size = len(nums) if len(nums) > 0 else 1
    ht = HashTable(size)
    
    for num in nums:
        if ht.get(num) is not None:
            return True
        ht.insert(num, True)
    return False

# Pengujian sesuai contoh di dokumen
print("Cek [1, 2, 3, 1]:", contains_duplicate([1, 2, 3, 1])) 
print("Cek [1, 2, 3, 4]:", contains_duplicate([1, 2, 3, 4])) 
print("Cek []:", contains_duplicate([]))                    
```

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(self.size)]
        self.keys_order = []

    def _hash(self, key):
        return sum(ord(char) for char in str(key)) % self.size

    def insert(self, key, value):
        index = self._hash(key)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value)
                return
        self.table[index].append((key, value))
        self.keys_order.append(key)

    def get(self, key):
        index = self._hash(key)
        for k, v in self.table[index]:
            if k == key:
                return v
        return None

    def delete(self, key):
        index = self._hash(key)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                del self.table[index][i]
                if key in self.keys_order:
                    self.keys_order.remove(key)
                return True
        return False

    def __str__(self):
        items = []
        for key in self.keys_order:
            val = self.get(key)
            if val is not None:
                items.append(f"{repr(key)}: {repr(val)}")
        return "{" + ", ".join(items) + "}"


def hitung_frekuensi_karakter(teks, abaikan=False):
    ht = HashTable(size=26)
    if abaikan:
        teks = teks.lower()
    for char in teks:
        if abaikan and not char.isalnum():
            continue
        current_count = ht.get(char)
        if current_count is None:
            ht.insert(char, 1)
        else:
            ht.insert(char, current_count + 1)
    return str(ht)

teks1 = "hello world"
print(f'"{teks1}" -> {hitung_frekuensi_karakter(teks1)}')

teks2 = "Python is Fun!"
print(f' {hitung_frekuensi_karakter(teks2)}')

```
