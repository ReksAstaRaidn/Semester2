# constructor class #

```python
class Node:
    def __init__(self, data=None, next_node=None):
        self.data = data
        self.next = next_node
```

# membuat fungsi untuk menampilkan data linked list #

```python
def print_linked_list(head):
    current = head
    while current is not None:
        print(current.data, end=" -> ")
        current = current.next
    print("None")
```

# Insert a value in linked list #

```python
def sorted_insert(head, new_node):
    
    # Special case for the empty linked list
    if head is None:
        new_node.next = head
        head = new_node
        
    # Special case for head at end
    elif head.data >= new_node.data:
        new_node.next = head
        head = new_node
        
    else:
        # Locate the node before the point of insertion
        current = head
        while current.next is not None and current.next.data < new_node.data:
            current = current.next
        new_node.next = current.next
        current.next = new_node
    return head
```

```python
# Test function
def test_sorted_insert():
    # Create linked list: 1 -> 3 -> 4 -> 7
    head = Node(1)
    node1 = Node(3)
    node2 = Node(4)
    node3 = Node(7)
    head.next = node1
    node1.next = node2
    node2.next = node3
    print_linked_list(head)
    
    # Insert 2: 1 -> 2 -> 3 -> 4 -> 7
    print('insert 2')
    new_node = Node(2)
    head = sorted_insert(head, new_node)
    print_linked_list(head)
    
    # Insert 0: 0 -> 1 -> 2 -> 3 -> 4 -> 7
    print('insert 0')
    new_node = Node(0)
    head = sorted_insert(head, new_node)
    print_linked_list(head)
    
    # Insert 8: 0 -> 1 -> 2 -> 3 -> 4 -> 7 -> 8
    print('insert 8')
    new_node = Node(8)
    head = sorted_insert(head, new_node)
    print_linked_list(head)
    
test_sorted_insert()
```

# Insert a node at the beginning of a Linked List #

```python
def insert_at_beginning(head, new_data):
    # 1. Buat node baru dengan data yang diberikan
    new_node = Node(new_data)
    # 2. Jadikan 'next' dari node baru menunjuk ke head saat ini
    new_node.next = head
    # 3. Ubah head menjadi node baru
    head = new_node
    return head
```

```python
def test_insert_at_beginning():
    # Test case 1: Linked list kosong
    print("\nTest Case 1: Insert into empty list")
    head1 = None
    head1 = insert_at_beginning(head1, 3)
    print_linked_list(head1) # Output: 3 -> None
    
    # Test case 2: Linked list tidak kosong
    print("\nTest Case 2: Insert into non-empty list")
    head2 = Node(1, Node(2, Node(3)))
    print("Original list:")
    print_linked_list(head2) # Output: 1 -> 2 -> 3 -> None
    head2 = insert_at_beginning(head2, 0)
    print("List after inserting 0 at beginning:")
    print_linked_list(head2) # Output: 0 -> 1 -> 2 -> 3 -> None
    
    # Test case 3: Insert lagi
    print("\nTest Case 3: Insert another value at beginning")
    head2 = insert_at_beginning(head2, -1)
    print("List after inserting -1 at beginning:")
    print_linked_list(head2) # Output: -1 -> 0 -> 1 -> 2 -> 3 -> None

test_insert_at_beginning()
```

# Insert a node at the end of a Linked List #

```python
def insert_at_end(head, new_data):
    
    # 1. Buat node baru
    new_node = Node(new_data)
    
    # 2. Jika linked list kosong, jadikan node baru sebagai head
    if head is None:
        return new_node
        
    # 3. Traversal hingga node terakhir
    current = head
    while current.next:
        current = current.next
        
    # 4. Jadikan 'next' dari node terakhir menunjuk ke node baru
    current.next = new_node
    
    return head
```

```python
def test_insert_at_end():
    # Test case 1: Linked list kosong
    print("\nTest Case 1: Insert into empty list")
    head1 = None
    head1 = insert_at_end(head1, 10)
    print_linked_list(head1) # Output: 10 -> None
    
    # Test case 2: Linked list tidak kosong
    print("\nTest Case 2: Insert into non-empty list")
    head2 = Node(1)
    print("Original list:")
    print_linked_list(head2) # Output: 1 -> None
    head2 = insert_at_end(head2, 2)
    print("List after inserting 2 at end:")
    print_linked_list(head2) # Output: 1 -> 2 -> None
    
    # Test case 3: Insert lagi di akhir
    print("\nTest Case 3: Insert another value at end")
    head2 = insert_at_end(head2, 3)
    print("List after inserting 3 at end:")
    print_linked_list(head2) # Output: 1 -> 2 -> 3 -> None
    
test_insert_at_end()
```

# Insert a node in the middle of a Linked List (after a given node) #

```python
def insert_after_node(prev_node, new_data):
    
    # 1. Periksa apakah prev_node ada
    if prev_node is None:
        print("Node sebelumnya tidak boleh None.")
        return
        
    # 2. Buat node baru
    new_node = Node(new_data)
    
    # 3. Jadikan 'next' dari node baru menunjuk ke 'next' dari prev_node
    new_node.next = prev_node.next
    
    # 4. Jadikan 'next' dari prev_node menunjuk ke node baru
    prev_node.next = new_node
```

```python
def test_insert_after_node():
    # Create a linked list: 1 -> 3 -> 4
    head = Node(1, Node(3, Node(4)))
    print("\nOriginal list:")
    print_linked_list(head) # Output: 1 -> 3 -> 4 -> None
    
    # Find the node with data 1 (which is the head)
    prev_node = head
    insert_after_node(prev_node, 2) # Insert 2 after node 1
    print("List after inserting 2 after node 1:")
    print_linked_list(head) # Output: 1 -> 2 -> 3 -> 4 -> None
    
    # Find the node with data 3
    current = head
    while current and current.data != 3:
        current = current.next
    prev_node = current
    insert_after_node(prev_node, 3.5) # Insert 3.5 after node 3
    print("List after inserting 3.5 after node 3:")
    print_linked_list(head) # Output: 1 -> 2 -> 3 -> 3.5 -> 4 -> None

    # Insert at the end using this function (insert after the last node)
    current = head
    while current.next:
        current = current.next
    prev_node = current
    insert_after_node(prev_node, 5) # Insert 5 after node 4
    print("List after inserting 5 after node 4 (at end):")
    print_linked_list(head) # Output: 1 -> 2 -> 3 -> 3.5 -> 4 -> 5 -> None
test_insert_after_node()
```

# delete a given node in Linked List under given constraints #

```python
def delete_node(head, data):
    # initialize both temp and prev as head
    temp = head
    prev = head
    # if node to be deleted is head node
    # if it is node just print the fact that head is empty
    # if node is not None we just replace head (temp) with next
    if temp.data == data:
        if temp.next is None:
            # list has only one node and we are removing it
            # so we return None
            return None
        else:
            temp.data = temp.next.data
            temp.next = temp.next.next
            return head
        
    # move forward prev and temp until:
    # 1. we reach to last node
    # 2. we found match between data
    while temp.next is not None and temp.data != data:
        prev = temp
        temp = temp.next
        
    # Three main cases:
    # 1. If temp is last value and we don't find match
    if temp.next is None and temp.data != data:
        print("Can't delete the node as it doesn't exist")
        return head
    # 2. If node is last node and we found match
    elif temp.next is None and temp.data == data:
        prev.next = None
        return head
    # 3. Temp is data we want to remove so we pointed prev next to
    # temp next --> removing temp
    else:
        prev.next = temp.next
        return head
```

```python
# Test case
head = Node(1, Node(2, Node(3, Node(4, Node(5)))))
print_linked_list(head) # 1 2 3 4 5

print('Removing 3')
head = delete_node(head, 3)
print_linked_list(head) # 1 2 4 5

print('Removing 1')
head = delete_node(head, 1)
print_linked_list(head) # 2 4 5

print('Removing 5')
head = delete_node(head, 5)
print_linked_list(head) # 2 4

print('Removing 6')
head = delete_node(head, 6)
print_linked_list(head) # Can't delete the node as it doesn't exist, 2 4

print('Removing 2')
head = delete_node(head, 2)
print_linked_list(head) # Can't delete the node as it doesn't exist, 2 4

print('Removing 4')
head = delete_node(head, 4)
print_linked_list(head) # Can't delete the node as it doesn't exist, 2 4
```

# Compare two strings represented as linked lists #

```python
def compare(list1, list2):
    
    # Traverse both lists. Stop when either end of linked
    # list is reached to end or current characters don't match
    while(list1 and list2 and list1.data == list2.data):
        list1 = list1.next
        list2 = list2.next
        
    # If both lists are not empty, compare mismatching
    # characters
    if(list1 and list2):
        return 1 if list1.data > list2.data else -1
        
    # if second string reach end but still string 1 has value
    if (list1 and not list2):
        return 1
    # if first string reach end but still string 2 has value
    if (list2 and not list1):
        return -1
    return 0
```

```python
# Test case 1: both lists are equal
print("abc vs abc - return 0")
list1 = Node('a')
list1.next = Node('b')
list1.next.next = Node('c')

list2 = Node('a')
list2.next = Node('b')
list2.next.next = Node('c')

if compare(list1, list2) == 0:
    print("test 1 is successfully passed")
    
# Test case 2: list1 is lexicographically greater
print("abc vs abb - return 1")
list1 = Node('a')
list1.next = Node('b')
list1.next.next = Node('c')

list2 = Node('a')
list2.next = Node('b')
list2.next.next = Node('b')

if compare(list1, list2) == 1:
    print("test 2 is successfully passed")
    
# Test case 3: list2 is lexicographically greater
print("abb vs abs - return -1")
list1 = Node('a')
list1.next = Node('b')
list1.next.next = Node('b')

list2 = Node('a')
list2.next = Node('b')
list2.next.next = Node('c')

if compare(list1, list2) == -1:
    print("test 3 is successfully passed")
```

# Reverse Linked List # 

```python
def reverse_linked_list(head):
    
    prev = None
    current = head
    
    while current:
        temp = current.next
        current.next = prev
        prev = current
        current = temp
        
    return prev
```

```python
def test_reverse_linked_list():
    # Create a linked list: 1 -> 2 -> 3 -> 4 -> 5
    head = Node(1)
    head.next = Node(2)
    head.next.next = Node(3)
    head.next.next.next = Node(4)
    head.next.next.next.next = Node(5)
    
    print(' \n Printing list')
    print_linked_list(head)
    # Reverse the linked list
    reversed_head = reverse_linked_list(head)
    
    print('\n Printing reversed list')
    # Print the reversed linked list: 5 -> 4 -> 3 -> 2 -> 1
    print_linked_list(reversed_head)
    
test_reverse_linked_list()
```
