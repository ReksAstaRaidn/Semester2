# Inisiasi Struktur Data Tree

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = [] # Setiap node dapat memiliki daftar anak

    def add_child(self, child_node):
        # Menambahkan anak ke daftar anak node ini
        self.children.append(child_node)

    def __str__(self, level=0):
        # Metode untuk mencetak representasi tree secara visual
        ret = " " * level + str(self.data) + "\n"
        for child in self.children:
            ret += child. __str__ (level + 1)
        return ret

# --- Contoh Penggunaan ---
print("Membuat Struktur Tree Sederhana:")

# Membuat node root
root = TreeNode("Global")

# Membuat node anak untuk root
region_asia = TreeNode("Asia")
region_europe = TreeNode("Europe")

root.add_child(region_asia)
root.add_child(region_europe)

# Membuat node anak untuk Asia
country_singapore = TreeNode("Singapore")
country_japan = TreeNode("Japan")

region_asia.add_child(country_singapore)
region_asia.add_child(country_japan)

# Membuat node anak untuk Europe
country_germany = TreeNode("Germany")
country_netherland = TreeNode("Netherland")

region_europe.add_child(country_germany)
region_europe.add_child(country_netherland)

# Mencetak tree
print(root)
```

# Menghapus Child

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = [] # Setiap node dapat memiliki daftar anak

    def add_child(self, child_node):
        # Menambahkan anak ke daftar anak node ini
        self.children.append(child_node)

    def remove_child(self, child_node):
        # Menghapus node anak tertentu dari daftar
        self.children = [child for child in self.children if child != child_node]

    def remove_by_data(self, data_to_remove):
        # mencari dan menghapus anak yang memiliki data tertentu 
        initial_count = len(self.children)
        self.children = [child for child in self.children if child.data != data_to_remove]

        # jika tidak ditemukan di level ini, cari level di bawahnya (rekursif)
        if len(self.children) == initial_count:
            for child in self.children:
                child.remove_by_data(data_to_remove)

    def __str__(self, level=0):
        # Metode untuk mencetak representasi tree secara visual
        ret = " " * level + str(self.data) + "\n"
        for child in self.children:
            ret += child. __str__ (level + 1)
        return ret

# Contoh penggunaan:
root = TreeNode("Root")
child1 = TreeNode("Anak 1")
child2 = TreeNode("Anak 2")

root.add_child(child1)
root.add_child(child2)

print("Sebelum dihapus:")
print(root)

root.remove_by_data("Anak 1")

print("Setelah 'Anak 1' dihapus:")
print(root)
```

# Binary Tree

```python
class BinaryNode:
    def __init__ (self, data):
        self.data = data
        self.left = None
        self.right = None

    def insert_left(self, data):
        if self.left is None:
            self.left = BinaryNode(data)
        else:
            new_node = BinaryNode(data)
            new_node.left = self.left
            self.left = new_node

    def insert_right(self, data):
        if self.right is None:
            self.right = BinaryNode(data)
        else:
            new_node = BinaryNode(data)
            new_node.right = self.right
            self.right = new_node

    def delete(self, data):
        # Mencari dan menghapus anak kiri jika datanya cocok
        if self.left and self.left.data == data:
            self.left = None
            return True
        # Mencari dan menghapus anak kanan jika datanya cocok
        if self.right and self.right.data == data:
            self.right = None
            return True
        # Rekursi ke bawah jika tidak ditemukan di level ini
        res = False
        if self.left:
            res = self.left.delete(data)
        if not res and self.right:
            res = self.right.delete(data)
        return res

    def __str__ (self, level=0, prefix="Root: "):
        ret = " " * level + prefix + str(self.data) + "\n"
        if self.left:
            ret += self.left.__str__ (level + 1, "L--- ")
        if self.right:
            ret += self.right.__str__ (level + 1, "R--- ")
        return ret

# --- Contoh Penggunaan ---
root = BinaryNode("A")
root.insert_left("B")
root.insert_right("C")
root.left.insert_left("D")

print("Tree awal:")
print(root)

print("Menghapus node 'B' (beserta seluruh cabangnya):")
root.delete("B")
print(root)
```

```python
class ExpNode:
    def __init__ (self, value):
        self.value = value
        self.left = None
        self.right = None

def evaluate(node):
    # Jika leaf node (angka), kembalikan nilainya
    if node.left is None and node.right is None:
        return int(node.value)

    left_val = evaluate(node.left)
    right_val = evaluate(node.right)

    # Terapkan operator
    if node.value == '+': return left_val + right_val
    if node.value == '-': return left_val - right_val
    if node.value == '*': return left_val * right_val
    if node.value == '/': return left_val / right_val

# Membangun pohon untuk (3 + 5) * 2
#      *
#   /    \
#   +    2
#  /  \
#  3  5

root_exp = ExpNode('*')
root_exp.right = ExpNode('2')
root_exp.left = ExpNode('+')
root_exp.left.left = ExpNode('3')
root_exp.left.right = ExpNode('5')

print("Ekspresi: (3 + 5) * 2")
print("Hasil evaluasi:", evaluate(root_exp))
```

# Implementasi Traversal

```python
# Diasumsikan kelas BinaryNode sudah didefinisikan sebelumnya
# from cell 'igUIxnHrKLv0'
def inorder_traversal(node): 
    """
    Traversasi Inorder (Left -> Root -> Right).
    Mengunjungi node kiri, lalu node saat ini, lalu node kanan.
    """
    if node:
        inorder_traversal(node.left)
        print(node.data, end=" ")
        inorder_traversal(node.right)

def preorder_traversal(node):
    """
    Traversasi Preorder (Root -> Left -> Right).
    Mengunjungi node saat ini, lalu node kiri, lalu node kanan.
    """
    if node:
        print(node.data, end=" ")
        preorder_traversal(node.left)
        preorder_traversal(node.right)

def postorder_traversal(node):
    """
    Traversasi Postorder (Left -> Right -> Root).
    Mengunjungi node kiri, lalu node kanan, lalu node saat ini.
    """
    if node:
        postorder_traversal(node.left)
        postorder_traversal(node.right)
        print(node.data, end=" ")

# --- Contoh Penggunaan ---
print("\n--- Contoh Penerapan DFS Traversal ---")
# Membuat sebuah Binary Tree:
#     1
#   /  \
#  2   3
# / \ / \
# 4 5 6 7

root_dfs = BinaryNode(1)
root_dfs.insert_left(2)
root_dfs.insert_right(3)
root_dfs.left.insert_left(4)
root_dfs.left.insert_right(5)
root_dfs.right.insert_left(6)
root_dfs.right.insert_right(7)

print("Struktur Binary Tree:")
print(root_dfs)
print("Traversasi Inorder:")
inorder_traversal(root_dfs)
print("\n")

print("Traversasi Preorder:")
preorder_traversal(root_dfs)
print("\n")

print("Traversasi Postorder:")
postorder_traversal(root_dfs)
print("\n")
```
