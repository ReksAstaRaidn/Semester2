# Antrian Pelanggan

```python
# Inisialisasi antrian kosong
customer_queue = []

# Fungsi untuk menambahkan elemen ke antrian (Enqueue)
def enqueue(queue, item):
    queue.append(item)
    print(f"Menambahkan '{item}' ke antrian. Antrian saat ini: {queue}")
    # Fungsi untuk menghapus elemen dari depan antrian (Dequeue)

def dequeue(queue):
    if not isEmpty(queue):
        item = queue.pop(0) # Mengambil elemen pertama (depan)
        print(f"Melayani '{item}' dari antrian. Antrian saat ini: {queue}")
        return item
    else:
        print("Antrian kosong, tidak ada yang bisa dilayani.")
        return None

# Fungsi untuk melihat elemen paling depan antrian (Front)
def front(queue):
    if not isEmpty(queue):
        return queue[0]
    else:
        print("Antrian kosong.")
        return None

# Fungsi untuk memeriksa apakah antrian kosong (isEmpty)
def isEmpty(queue):
    return len(queue) == 0

# Fungsi untuk mendapatkan ukuran antrian (Size)
def size(queue):
    return len(queue)

# --- Simulasi Antrian Pelanggan ---
print("\n--- Simulasi Antrian Pelanggan ---")

enqueue(customer_queue, "Pelanggan A")
enqueue(customer_queue, "Pelanggan B")
enqueue(customer_queue, "Pelanggan C")

print(f"\nAntrian saat ini: {customer_queue}")
print(f"Pelanggan paling depan: {front(customer_queue)}")
print(f"Jumlah pelanggan dalam antrian: {size(customer_queue)}")

dequeue(customer_queue)
dequeue(customer_queue)

print(f"\nAntrian setelah 2 pelanggan dilayani: {customer_queue}")
print(f"Apakah antrian kosong? {isEmpty(customer_queue)}")

enqueue(customer_queue, "Pelanggan D")

print(f"\nAntrian saat ini: {customer_queue}")
print(f"Pelanggan paling depan: {front(customer_queue)}")
print(f"Jumlah pelanggan dalam antrian: {size(customer_queue)}")

dequeue(customer_queue)
dequeue(customer_queue)
dequeue(customer_queue) # Coba dequeue saat antrian kosong
```

# Studi Kasus: Antrian Tugas Cek (Print Job Queue)

```python
from collections import deque

# Inisialisasi antrian tugas cetak kosong menggunakan deque
print_queue = deque()

# Fungsi untuk menambahkan tugas cetak ke antrian
def add_print_job(queue, job_name):
    queue.append(job_name)
    print(f"Menambahkan '{job_name}' ke antrian. Antrian saat ini: {list(queue)}")

# Fungsi untuk memproses (menghapus) tugas cetak dari depan antrian
def process_next_job(queue):
    if not is_queue_empty(queue):
        job = queue.popleft() # Mengambil elemen pertama dari depan (lebih efisien dengan deque)
        print(f"Memproses '{job}' dari antrian. Antrian saat ini: {list(queue)}")
        return job
    else:
        print("Antrian cetak kosong, tidak ada tugas untuk diproses.")
        return None

# Fungsi untuk melihat tugas cetak paling depan antrian
def peek_next_job(queue):
    if not is_queue_empty(queue):
        return queue[0]
    else:
        print("Antrian cetak kosong.")
        return None

# Fungsi untuk memeriksa apakah antrian kosong
def is_queue_empty(queue):
    return len(queue) == 0

# Fungsi untuk mendapatkan ukuran antrian
def get_queue_size(queue):
    return len(queue)

# --- Simulasi Antrian Tugas Cetak ---
print("\n--- Simulasi Antrian Tugas Cetak ---")

add_print_job(print_queue, "Dokumen Laporan Keuangan")
add_print_job(print_queue, "Gambar Desain Proyek")
add_print_job(print_queue, "Tabel Data Penjualan")

print(f"\nAntrian cetak saat ini: {list(print_queue)}" )
print(f"Tugas cetak berikutnya: {peek_next_job(print_queue)}")
print(f"Jumlah tugas dalam antrian: {get_queue_size(print_queue)}")

process_next_job(print_queue)
process_next_job(print_queue)

print(f"\nAntrian setelah 2 tugas diproses: {list(print_queue)}")
print(f"Apakah antrian kosong? {is_queue_empty(print_queue)}")

add_print_job(print_queue, "Presentasi Rapat")

print(f"\nAntrian cetak saat ini: {list(print_queue)}")
print(f"Tugas cetak berikutnya: {peek_next_job(print_queue)}")
print(f"Jumlah tugas dalam antrian: {get_queue_size(print_queue)}")

process_next_job(print_queue)
process_next_job(print_queue)
process_next_job(print_queue) # Coba proses saat antrian kosong
```

# Implementasi Queue menggunakan Linked List

```python
# 1. Mendefinisikan kelas Node
class Node:
    def __init__ (self, data):
        self.data = data
        self.next = None # Pointer ke node berikutnya
```

```python
# 2. Mendefinisikan kelas Queue menggunakan Linked List
class LinkedListQueue:
    def __init__ (self):
        self.front = None # Node paling depan antrian
        self.rear = None # Node paling belakang antrian
        self.count = 0 # Jumlah elemen dalam antrian
        
    # Operasi Enqueue: Menambahkan elemen ke belakang antrian
    def enqueue(self, item):
        new_node = Node(item)
        if self.is_empty():
            self.front = new_node
            self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node
            
        self.count += 1
        print(f"Menambahkan '{item}' ke antrian. Antrian saat ini: {self._get_queue_elements()}")

    # Operasi Dequeue: Menghapus elemen dari depan antrian
    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong, tidak ada yang bisa dilayani.")
            return None
        item = self.front.data
        self.front = self.front.next
        if self.front is None: # Jika antrian menjadi kosong setelah dequeue
            self.rear = None
        self.count -= 1
        print(f"Melayani '{item}' dari antrian. Antrian saat ini: {self._get_queue_elements()}")
        return item
    
    # Operasi Front: Melihat elemen paling depan antrian tanpa menghapusnya
    def front_item(self):
        if self.is_empty():
            print("Antrian kosong.")
            return None
        return self.front.data

    # Operasi Front: Melihat elemen paling depan antrian tanpa menghapusnya
    def front_item(self):
        if self.is_empty():
            print("Antrian kosong.")
            return None
        return self.front.data

    # Operasi Is_Empty: Memeriksa apakah antrian kosong
    def is_empty(self):
        return self.front is None

    # Operasi Size: Mendapatkan jumlah elemen dalam antrian
    def size(self):
        return self.count

    # Helper function untuk menampilkan elemen antrian (untuk debugging/visualisasi)
    def _get_queue_elements(self):
        elements = []
        current = self.front
        while current:
            elements.append(current.data)
            current = current.next
        return elements

# --- Simulasi Antrian ---
print("\n--- Simulasi Antrian Menggunakan Linked List ---")
queue_ll = LinkedListQueue()

queue_ll.enqueue("Pesanan A")
queue_ll.enqueue("Pesanan B")
queue_ll.enqueue("Pesanan C")

print(f"\nAntrian saat ini: {queue_ll._get_queue_elements()}")
print(f"Elemen paling depan: {queue_ll.front_item()}")
print(f"Jumlah elemen dalam antrian: {queue_ll.size()}")

queue_ll.dequeue()
queue_ll.dequeue()

print(f"\nAntrian setelah 2 pesanan dilayani: {queue_ll._get_queue_elements()}")
print(f"Apakah antrian kosong? {queue_ll.is_empty()}")

queue_ll.enqueue("Pesanan D")

print(f"\nAntrian saat ini: {queue_ll._get_queue_elements()}")
print(f"Elemen paling depan: {queue_ll.front_item()}")
print(f"Jumlah elemen dalam antrian: {queue_ll.size()}")

queue_ll.dequeue()
queue_ll.dequeue()
queue_ll.dequeue() # Coba dequeue saat antrian kosong
```
