```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = [] # Setiap node dapat memiliki daftar anak

    def add_child(self, child_node):
        # Menambahkan anak ke daftar anak node ini
        self.children.append(child_node)

    def remove_child(self, child_node):
        # Menghapus node anak tertentu dari daftar
        self.children = [child for child in self.children if child != child_node]

    def remove_by_data(self, data_to_remove):
        # mencari dan menghapus anak yang memiliki data tertentu 
        initial_count = len(self.children)
        self.children = [child for child in self.children if child.data != data_to_remove]

        # jika tidak ditemukan di level ini, cari level di bawahnya (rekursif)
        if len(self.children) == initial_count:
            for child in self.children:
                child.remove_by_data(data_to_remove)

    def __str__(self, level=0):
        # Metode untuk mencetak representasi tree secara visual
        ret = " " * level + str(self.data) + "\n"
        for child in self.children:
            ret += child. __str__ (level + 1)
        return ret
```

```python
class BinaryNode:
    def __init__ (self, data):
        self.data = data
        self.left = None
        self.right = None

    def insert_left(self, data):
        if self.left is None:
            self.left = BinaryNode(data)
        else:
            new_node = BinaryNode(data)
            new_node.left = self.left
            self.left = new_node

    def insert_right(self, data):
        if self.right is None:
            self.right = BinaryNode(data)
        else:
            new_node = BinaryNode(data)
            new_node.right = self.right
            self.right = new_node

    def delete(self, data):
        # Mencari dan menghapus anak kiri jika datanya cocok
        if self.left and self.left.data == data:
            self.left = None
            return True
        # Mencari dan menghapus anak kanan jika datanya cocok
        if self.right and self.right.data == data:
            self.right = None
            return True
        # Rekursi ke bawah jika tidak ditemukan di level ini
        res = False
        if self.left:
            res = self.left.delete(data)
        if not res and self.right:
            res = self.right.delete(data)
        return res

    def __str__ (self, level=0, prefix="Root: "):
        ret = " " * level + prefix + str(self.data) + "\n"
        if self.left:
            ret += self.left.__str__(level + 1, "L--- ")
        if self.right:
            ret += self.right.__str__(level + 1, "R--- ")
        return ret
```

```python
root = BinaryNode("0")

root.insert_left("1")
root.insert_right("2")

root.left.insert_left("3")
root.left.insert_right("4")
root.right.insert_left("5")
root.right.insert_right("6")

root.left.left.insert_left("7")
root.left.left.insert_right("8")

root.left.right.insert_left("9")
root.left.right.insert_right("10")

root.right.left.insert_left("11")
root.right.left.insert_right("12")

root.right.right.insert_left("13")
root.right.right.insert_right("14")



print(root)
```
