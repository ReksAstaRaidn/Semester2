```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class Queue:
    def __init__(self, max_size=10):
        self.head = None
        self.tail = None
        self.size = 0
        self.max_size = max_size
 
    def is_empty(self):
        return self.size == 0
 
    def is_full(self):
        return self.size == self.max_size
 
    def enqueue(self, data):
        if self.is_full():
            print(f"Antrian penuh, Maksimum {self.max_size} pesanan.")
            return False
        node = Node(data)
        if self.tail:
            self.tail.next = node
        self.tail = node
        if self.head is None:
            self.head = node
        self.size += 1
        return True
 
    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong! Tidak ada pesanan yang bisa diproses.")
            return None
        data = self.head.data
        self.head = self.head.next
        if self.head is None:
            self.tail = None
        self.size -= 1
        return data
 
    def display(self):
        if self.is_empty():
            print("  (kosong)")
            return
        current = self.head
        items = []
        while current:
            items.append(current.data)
            current = current.next
        print("  " + " -> ".join(f"[{item}]" for item in items) + "  (depan -> belakang)")
```

```python
class Stack:
    def __init__(self):
        self.top = None
        self.size = 0
 
    def is_empty(self):
        return self.size == 0
 
    def push(self, data):
        node = Node(data)
        node.next = self.top
        self.top = node
        self.size += 1
 
    def pop(self):
        if self.is_empty():
            print("Riwayat kosong! Tidak ada pesanan untuk di-undo.")
            return None
        data = self.top.data
        self.top = self.top.next
        self.size -= 1
        return data
 
    def display(self):
        if self.is_empty():
            print("  (kosong)")
            return
        current = self.top
        items = []
        while current:
            items.append(current.data)
            current = current.next
        print("  " + " -> ".join(f"[{item}]" for item in items) + "  (atas -> bawah)")
```

```python
def printStatus(queue, stack):
    print("\nStatus Antrian:")
    queue.display()
    print("Riwayat Pesanan:")
    stack.display()
```

```python
def main():
    q = Queue(max_size=5)
    s = Stack()
    penghitungPesanan = 1
    
    while True:
        print("\nMenu:")
        print("1. Tambah Pesanan")
        print("2. Proses Pesanan")
        print("3. Undo Proses Pesanan")
        print("4. Tampilkan Status Antrian & Riwayat")
        print("5. Keluar")
        
        pilihan = input("Pilih menu (1-5): ")
        
        if pilihan == '1':
            namaBarang = input("Masukkan nama barang untuk pesanan baru: ")
            
            if not namaBarang:
                print("Nama barang tidak boleh kosong!")
                continue
            infoPesanan = f"Pesanan{penghitungPesanan}: {namaBarang}"
        
            if q.enqueue(infoPesanan):
                print(f"Pesanan '{infoPesanan}' berhasil ditambahkan ke antrian.")
                penghitungPesanan += 1
        
        elif pilihan == '2':        
            pesanan = q.dequeue()
            if pesanan: 
                s.push(pesanan)
                print(f"Memproses {pesanan}...")
        
        elif pilihan == '3':
            pesanan = s.pop()
            if pesanan:
                q.enqueue(pesanan)
                print(f"Pesanan '{pesanan}' dikembalikan ke antrian untuk diproses ulang.")
        
        elif pilihan == '4':
            printStatus(q, s)   
        
        elif pilihan == '5':
            print("Terima kasih")
            break
        else:
            print("Pilihan tidak valid. Silakan pilih menu 1-5.")
```

```python
if __name__ == "__main__":
    main()
```
