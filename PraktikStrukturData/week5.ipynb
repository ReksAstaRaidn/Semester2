# Implementasi Double Linked List #

```python
class Node:
    """Represents a single node in a doubly linked list."""
    def __init__ (self, data):
        self.data = data
        self.next = None
        self.prev = None
```

```python
class DoublyLinkedList:
    """Implements a doubly linked list with basic operations."""
    def __init__ (self):
        self.head = None
        self.tail = None
        
    def is_empty(self):
        """Checks if the list is empty."""
        return self.head is None
        
    def append(self, data):
        """Adds an item to the end of the list."""
        new_node = Node(data)
        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node

    def prepend(self, data):
        """Adds an item to the beginning of the list."""
        new_node = Node(data)
        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            self.head.prev = new_node
            new_node.next = self.head
            self.head = new_node

    def display_forward(self):
        """Prints the list from head to tail."""
        current = self.head
        result = []
        while current:
            result.append(str(current.data))
            current = current.next
        print("Forward: <-->".join(result))

    def display_backward(self):
        """Prints the list from tail to head."""
        current = self.tail
        result = []
        while current:
            result.append(str(current.data))
            current = current.prev
        print("Backward: <-->".join(result))

    def find(self, data):
        """Searches for a node with the given data and returns it."""
        current = self.head
        while current:
            if current.data == data:
                return current
            current = current.next
        return None

    def delete(self, data):
        """Deletes the first node with the given data from the list."""
        if self.is_empty():
            print(f"Daftar kosong, tidak bisa menghapus {data}.")
        return

        node_to_delete = self.find(data)
        if not node_to_delete:
            print(f"Node dengan nilai {data} tidak ditemukan.")
        return
        
        # If it's the head node
        if node_to_delete.prev is None:
            self.head = node_to_delete.next
        else:
            node_to_delete.prev.next = node_to_delete.next
        
        # If it's the tail node
        if node_to_delete.next is None:
            self.tail = node_to_delete.prev
        else:
            node_to_delete.next.prev = node_to_delete.prev
            
        # If the list becomes empty after deletion
        if not self.head:
            self.tail = None
        elif not self.tail:
            self.head = None # This case should ideally not happen if head is not None already handled
        print(f"Node dengan nilai {data} berhasil dihapus.")
```

```python
# Contoh penerapan:
dll = DoublyLinkedList()
dll.append(1)
dll.append(2)
dll.prepend(0)

print("Doubly Linked List Contents:")
dll.display_forward()
dll.display_backward()
```

```python
# Contoh mencari node (tetap ada)
node_to_find = 2
found_node = dll.find(node_to_find)

if found_node:
    print(f"Node dengan nilai {node_to_find} ditemukan: {found_node.data}")
    if found_node.prev:
        print(f" Node sebelumnya: {found_node.prev.data}")
    else:
        print(" Ini adalah head node.")
    if found_node.next:
        print(f" Node berikutnya: {found_node.next.data}")
    else:
        print(" Ini adalah tail node.")
else:
    print(f"Node dengan nilai {node_to_find} tidak ditemukan.")
    
node_to_find_non_existent = 5
found_node_non_existent = dll.find(node_to_find_non_existent)

if found_node_non_existent:
    print(f"Node dengan nilai {node_to_find_non_existent} ditemukan.")
else:
    print(f"Node dengan nilai {node_to_find_non_existent} tidak ditemukan.")
```

```python
class BrowserHistory:
    """Implements a browser history using a Doubly Linked List."""
    def __init__(self, homepage):
        self.current = Node(homepage)
        self.head = self.current
        self.tail = self.current
        
    def visit(self, url):
        """Visits a new URL, adding it to the history and clearing forward history."""
        new_node = Node(url)
        self.current.next = new_node
        new_node.prev = self.current
        self.current = new_node
        self.tail = self.current # New visit makes this the new tail
    
    def back(self, steps):
        """Goes back a specified number of steps in history."""
        for _ in range(steps):
            if self.current.prev:
                self.current = self.current.prev
            else:
                break # Cannot go back further
        return self.current.data
    
    def forward(self, steps):
        """Goes forward a specified number of steps in history."""
        for _ in range(steps):
            if self.current.next:
                self.current = self.current.next
            else:
                break # Cannot go forward further
        return self.current.data
    
    def display_history(self):
        """Displays the entire browser history, marking the current page."""
        current_node = self.head
        history_list = []
        while current_node:
            prefix = "" if current_node != self.current else "* "
            history_list.append(prefix + current_node.data)
            current_node = current_node.next
        print("\nRiwayat Penelusuran (Current Page ditandai dengan *):\n" + "\n".join(history_list))
```

```python
# Inisialisasi riwayat dengan halaman awal
browser = BrowserHistory("homepage.com")
browser.display_history()

# Mengunjungi beberapa halaman
browser.visit("google.com")
browser.visit("github.com")
browser.visit("stackoverflow.com")
browser.display_history()

# Kembali 2 langkah
print(f"\nKembali 2 langkah: {browser.back(2)}")
browser.display_history()

# Mengunjungi halaman baru dari posisi saat ini (menghapus riwayat ke depan)
browser.visit("new_search.com")
browser.display_history()

# Mencoba maju, tapi tidak ada riwayat maju
print(f"\nMaju 1 langkah: {browser.forward(1)}")
browser.display_history()

# Kembali lagi, lalu maju
print(f"\nKembali 1 langkah: {browser.back(1)}")
browser.display_history()
print(f"\nMaju 1 langkah: {browser.forward(1)}")
browser.display_history()
```
