# Adjency List

```python
import collections

class Graph:
    def __init__(self, directed=False):
        self.graph = collections.defaultdict(list)
        self.directed = directed

    def add_edge(self, u, v):
        self.graph[u].append(v)
        if not self.directed:
            self.graph[v].append(u)  # Untuk graph tak berarah, tambahkan juga edge dari v ke u

    def get_neighbors(self, u):
        return self.graph[u]

    def display(self):
        print("Adjacency List:")
        for node in sorted(self.graph.keys()):
            print(f"{node}: {self.graph[node]}")

# Membuat dan menampilkan graph tak berarah
print("Graph Tak Berarah:")
undirected_graph = Graph(directed=False)
undirected_graph.add_edge('A', 'B')
undirected_graph.add_edge('A', 'C')
undirected_graph.add_edge('B', 'D')
undirected_graph.add_edge('C', 'E')
undirected_graph.add_edge('D', 'E')
undirected_graph.display()

print("\nGraph Berarah:")
# Membuat dan menampilkan graph berarah
directed_graph = Graph(directed=True)
directed_graph.add_edge('0', '1')
directed_graph.add_edge('0', '2')
directed_graph.add_edge('1', '2')
directed_graph.add_edge('2', '0')
directed_graph.add_edge('2', '3')
directed_graph.add_edge('3', '3')
directed_graph.display()
```

# Adjency Matrix

```python
class AdjacencyMatrixGraph:
    def __init__(self, num_vertices, directed=False):
        self.num_vertices = num_vertices
        self.directed = directed
        # Inisialisasi matriks dengan semua nilai 0
        self.matrix = [[0] * num_vertices for _ in range(num_vertices)]
        self.vertex_map = {}
        self.reverse_vertex_map = []
        self.next_vertex_id = 0

    def _get_vertex_index(self, vertex_label):
        if vertex_label not in self.vertex_map:
            self.vertex_map[vertex_label] = self.next_vertex_id
            self.reverse_vertex_map.append(vertex_label)
            self.next_vertex_id += 1
        return self.vertex_map[vertex_label]

    def add_edge(self, u, v, weight=1):
        u_idx = self._get_vertex_index(u)
        v_idx = self._get_vertex_index(v)
        if u_idx >= self.num_vertices or v_idx >= self.num_vertices:
            print("Error: Vertex index out of bounds. Increase num_vertices or choose existing ones.")
            return
        self.matrix[u_idx][v_idx] = weight
        if not self.directed:
            self.matrix[v_idx][u_idx] = weight

    def display(self):
        print("Adjacency Matrix:")
        # Print header row
        header = [str(label) for label in self.reverse_vertex_map]
        print("  " + " ".join(header))
        for i, row in enumerate(self.matrix):
            if i < len(self.reverse_vertex_map):
                print(f"{self.reverse_vertex_map[i]} {' '.join(map(str, row))}")
            else:
                print(f"? {' '.join(map(str, row))}")

# Membuat dan menampilkan graph tak berarah menggunakan Adjacency Matrix
print("Graph Tak Berarah (Adjacency Matrix):")
undirected_adj_matrix = AdjacencyMatrixGraph(num_vertices=5, directed=False)
undirected_adj_matrix.add_edge('A', 'B')
undirected_adj_matrix.add_edge('A', 'C')
undirected_adj_matrix.add_edge('B', 'D')
undirected_adj_matrix.add_edge('C', 'E')
undirected_adj_matrix.add_edge('D', 'E')
undirected_adj_matrix.display()

print("\nGraph Berarah (Adjacency Matrix):")
directed_adj_matrix = AdjacencyMatrixGraph(num_vertices=4, directed=True)
directed_adj_matrix.add_edge('0', '1')
directed_adj_matrix.add_edge('0', '2')
directed_adj_matrix.add_edge('1', '2')
directed_adj_matrix.add_edge('2', '0')
directed_adj_matrix.add_edge('2', '3')
directed_adj_matrix.add_edge('3', '3')
directed_adj_matrix.display()
```

# BFS (Breadth First Search)

```python
import collections

def bfs(graph_obj, start_node):
    # Set untuk menyimpan node yang sudah dikunjungi
    visited = set()
    # Queue untuk BFS
    queue = collections.deque([start_node])
    # List untuk menyimpan urutan penjelajahan
    traversal_order = []

    visited.add(start_node)

    while queue:
        current_node = queue.popleft()
        traversal_order.append(current_node)

        # Jelajahi semua tetangga dari node saat ini
        for neighbor in graph_obj.get_neighbors(current_node):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return traversal_order

# Gunakan graph tak berarah yang sudah dibuat sebelumnya
print("BFS Traversal (Graph Tak Berarah dari A):")
bfs_result_undirected = bfs(undirected_graph, 'A')
print(bfs_result_undirected)

# Gunakan graph berarah yang sudah dibuat sebelumnya
print("\nBFS Traversal (Graph Berarah dari 2):")
bfs_result_directed = bfs(directed_graph, '2')
print(bfs_result_directed)
```

# DFS (Depth First Search)

```python
import collections

def dfs(graph_obj, start_node):
    # Set untuk menyimpan node yang sudah dikunjungi
    visited = set()
    # Stack untuk DFS (menggunakan list sebagai stack, pop() untuk mengambil elemen terakhir)
    stack = [start_node]
    # List untuk menyimpan urutan penjelajahan
    traversal_order = []

    while stack:
        current_node = stack.pop()
        if current_node not in visited:
            visited.add(current_node)
            traversal_order.append(current_node)

            # Jelajahi semua tetangga dari node saat ini
            # Di sini, kita akan menambahkan tetangga ke stack dalam urutan terbalik
            # agar node dengan label 'lebih kecil' (berdasarkan sorting) dieksplorasi duluan
            # jika kita pop dari stack.
            for neighbor in sorted(graph_obj.get_neighbors(current_node), reverse=True):
                if neighbor not in visited:
                    stack.append(neighbor)

    return traversal_order

# Gunakan graph tak berarah yang sudah dibuat sebelumnya
print("DFS Traversal (Graph Tak Berarah dari A):")
dfs_result_undirected = dfs(undirected_graph, 'A')
print(dfs_result_undirected)

# Gunakan graph berarah yang sudah dibuat sebelumnya
print("\nDFS Traversal (Graph Berarah dari 2):")
dfs_result_directed = dfs(directed_graph, '2')
print(dfs_result_directed)
```

```python
import sys 

# Install graphviz if not already installed. this is a common requirement for visualization.
# we check if it's already installed to avoid unnecessary re-installation.

try:
    import graphviz
except ImportError:
    print("Graphviz library not found. Installing...")
    !{sys.executable} -m pip install graphviz 
    import graphviz

# Function to visualize Graph (Adjacency List)
def visualize_adjacency_list(graph_obj, graph_name="Graph", highlight_path=None):
    """
    Visualizes a graph represented by an adjacency list using Graphviz.

    Args:
        graph_obj (Graph): The graph object (Adjacency List).
        graph_name (str): The name for the graph visualization.
        highlight_path (list): An optional list of nodes to highlight as a path.
    """
    dot = graphviz.Digraph(comment=graph_name) if graph_obj.directed else graphviz.Graph(comment=graph_name)
    dot.attr(rankdir='LR')  # Left to Right layout

    nodes_in_graph = set()
    for u, neighbors in graph_obj.graph.items():
        nodes_in_graph.add(str(u))
        for v in neighbors:
            nodes_in_graph.add(str(v))

    # Add all nodes first
    for node in sorted(list(nodes_in_graph)):
        dot.node(node, style='filled', fillcolor='lightblue')

    # Add edges
    added_edges = set()  # To avoid duplicate edges for undirected graph
    for u, neighbors in graph_obj.graph.items():
        for v in neighbors:
            # For undirected graphs, add edge only once (e.g., A-B, not B-A again)
            if not graph_obj.directed:
                if (str(v), str(u)) in added_edges or (str(u), str(v)) in added_edges:
                    continue
            dot.edge(str(u), str(v))
            added_edges.add((str(u), str(v)))

    # Highlight path if provided
    if highlight_path:
        for i in range(len(highlight_path)):
            node = str(highlight_path[i])
            # Highlight start node in red, other path nodes in lightgreen
            dot.node(node, style='filled',
                     fillcolor='red' if i == 0 else 'lightgreen',
                     fontcolor='white' if i == 0 else 'black')
            if i < len(highlight_path) - 1:
                u = str(highlight_path[i])
                v = str(highlight_path[i + 1])
                # Highlight path edges in red with thicker penwidth
                dot.edge(u, v, color='red', penwidth='2.0')

    return dot


# Function to visualize Adjacency Matrix
def visualize_adjacency_matrix(adj_matrix_obj, graph_name="Graph", highlight_path=None):
    """
    Visualizes a graph represented by an adjacency matrix using Graphviz.

    Args:
        adj_matrix_obj (AdjacencyMatrixGraph): The graph object (Adjacency Matrix).
        graph_name (str): The name for the graph visualization.
        highlight_path (list): An optional list of nodes to highlight as a path.
    """
    dot = graphviz.Digraph(comment=graph_name) if adj_matrix_obj.directed else graphviz.Graph(comment=graph_name)
    dot.attr(rankdir='LR')

    num_vertices = adj_matrix_obj.num_vertices
    # Map vertex indices back to their original labels for visualization
    vertex_labels = [
        adj_matrix_obj.reverse_vertex_map[i] if i < len(adj_matrix_obj.reverse_vertex_map) else str(i)
        for i in range(num_vertices)
    ]

    # Add all nodes
    for label in vertex_labels:
        dot.node(label, style='filled', fillcolor='lightblue')

    # Add edges
    added_edges = set()
    for i in range(num_vertices):
        for j in range(num_vertices):
            if adj_matrix_obj.matrix[i][j] != 0:
                u_label = vertex_labels[i]
                v_label = vertex_labels[j]
                weight = adj_matrix_obj.matrix[i][j]
                if not adj_matrix_obj.directed:
                    if (v_label, u_label) in added_edges or (u_label, v_label) in added_edges:
                        continue
                dot.edge(u_label, v_label, label=str(weight) if weight != 1 else '')
                added_edges.add((u_label, v_label))

    # Highlight path if provided
    if highlight_path:
        for i in range(len(highlight_path)):
            node = str(highlight_path[i])
            dot.node(node, style='filled',
                     fillcolor='red' if i == 0 else 'lightgreen',
                     fontcolor='white' if i == 0 else 'black')
            if i < len(highlight_path) - 1:
                u = str(highlight_path[i])
                v = str(highlight_path[i + 1])
                dot.edge(u, v, color='red', penwidth='2.0')

    return dot

# --- Visualisasi Graph Adjacency List ---
print("### Visualisasi Graph Menggunakan Adjacency List\n")

print("#### Graph Tak Berarah (Adjacency List):\n")
display(visualize_adjacency_list(undirected_graph, graph_name="Undirected Graph (Adjacency List)"))

print("\n#### Graph Berarah (Adjacency List):\n")
display(visualize_adjacency_list(directed_graph, graph_name="Directed Graph (Adjacency List)"))

print("\n### Visualisasi Traversal (BFS) untuk Adjacency List\n")
print("#### BFS Traversal (Graph Tak Berarah dari A):\n")
display(visualize_adjacency_list(undirected_graph, graph_name="BFS Undirected from A",
                                 highlight_path=bfs_result_undirected))

print("\n#### BFS Traversal (Graph Berarah dari 2):\n")
display(visualize_adjacency_list(directed_graph, graph_name="BFS Directed from 2",
                                 highlight_path=bfs_result_directed))

print("\n### Visualisasi Traversal (DFS) untuk Adjacency List\n")
print("#### DFS Traversal (Graph Tak Berarah dari A):\n")
display(visualize_adjacency_list(undirected_graph, graph_name="DFS Undirected from A",
                                 highlight_path=dfs_result_undirected))

print("\n#### DFS Traversal (Graph Berarah dari 2):\n")
display(visualize_adjacency_list(directed_graph, graph_name="DFS Directed from 2",
                                 highlight_path=dfs_result_directed))

```

# Dijkstra

```python
import heapq
import collections

class WeightedGraph:
    def __init__(self):
        # Graph disimpan sebagai adjacency list, dengan setiap edge memiliki bobot
        # self.graph[u] = [(v1, weight1), (v2, weight2), ...]
        self.graph = collections.defaultdict(list)
        self.nodes = set()  # Set untuk menyimpan semua node unik dalam graph

    def add_edge(self, u, v, weight):
        # Tambahkan edge dari u ke v dengan bobot
        self.graph[u].append((v, weight))
        self.nodes.add(u)  # Tambahkan u ke set node
        self.nodes.add(v)  # Tambahkan v ke set node

    def dijkstra(self, start_node):
        # Dictionary untuk menyimpan jarak terpendek dari start_node ke setiap node
        distances = {node: float('infinity') for node in self.nodes}
        distances[start_node] = 0

        # Priority queue untuk menyimpan (jarak, node)
        priority_queue = [(0, start_node)]

        # Dictionary untuk menyimpan jalur terpendek (previous node dalam jalur)
        previous_nodes = {}

        while priority_queue:
            current_distance, current_node = heapq.heappop(priority_queue)

            # Jika jarak yang diambil dari queue lebih besar dari jarak yang sudah diketahui,
            # berarti kita sudah menemukan jalur yang lebih pendek sebelumnya, jadi abaikan.
            if current_distance > distances[current_node]:
                continue

            # Jelajahi setiap tetangga dari current_node
            for neighbor, weight in self.graph[current_node]:
                distance = current_distance + weight

                # Jika jarak baru lebih pendek dari jarak yang sudah diketahui ke tetangga
                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    previous_nodes[neighbor] = current_node
                    heapq.heappush(priority_queue, (distance, neighbor))

        return distances, previous_nodes

    def get_shortest_path(self, start_node, end_node, previous_nodes):
        path = collections.deque()
        current = end_node
        while current is not None:
            path.appendleft(current)
            current = previous_nodes.get(current)
            if current == start_node:
                path.appendleft(current)
                break

        # If current node is not start_node and not in previous_nodes, there is no path
        if current is not None and current not in previous_nodes and current != start_node:
            return []  # No path found

        # Check if start_node is the first element of the path
        if path and path[0] == start_node:
            return list(path)
        else:
            return []  # No path found or start_node is not in path


# Membuat instance WeightedGraph
weighted_graph = WeightedGraph()

# Menambahkan edge dengan bobot
weighted_graph.add_edge('A', 'B', 4)
weighted_graph.add_edge('A', 'C', 2)
weighted_graph.add_edge('B', 'E', 3)
weighted_graph.add_edge('C', 'D', 2)
weighted_graph.add_edge('C', 'F', 4)
weighted_graph.add_edge('D', 'E', 3)
weighted_graph.add_edge('D', 'F', 1)
weighted_graph.add_edge('E', 'G', 2)
weighted_graph.add_edge('F', 'G', 1)

start_node_dijkstra = 'A'

# Menjalankan algoritma Dijkstra
distances, previous_nodes = weighted_graph.dijkstra(start_node_dijkstra)

print(f"Jarak terpendek dari node '{start_node_dijkstra}':")
for node, dist in distances.items():
    print(f"  Ke {node}: {dist}")

print(f"\nJalur terpendek dari node '{start_node_dijkstra}':")
for node in weighted_graph.nodes:
    if node != start_node_dijkstra:
        path = weighted_graph.get_shortest_path(start_node_dijkstra, node, previous_nodes)
        if path:
            print(f"  Ke {node}: {' -> '.join(path)}")
        else:
            print(f"  Ke {node}: Tidak ada jalur")
```
